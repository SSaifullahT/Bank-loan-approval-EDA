```python
import sys
#Appraoch -  Dividing the columns of dataframe into categorical and non categorical dataset. We are checking the 
#no of unique values in each col.If the number of unique values is less then 15 we consider it as categorical data
#otherwise non categorical. we are also removing non categorical dataset which doesnt have numerical values


def convert_cat_noncat(dff1):
    col_list_app=dff1.columns.values
    col_noncategorical_unique={}
    col_cat_unique= {}
    for x in col_list_app:
        y=(len(dff1[x].unique()))
        if y < 15 :
            col_cat_unique[x]=y
        else:
            col_noncategorical_unique[x]=y

    #removing non arthimatic non categorical data set \

    df_all =dff1
    for x in col_cat_unique:
        df_all[x]= pd.Categorical(df_all[x])
        df_all['code']= df_all[x].cat.codes
        df_all[x]=(df_all['code'])
    return df_all


```

```python
#segmenting data Client with payment difficulties i.e. TARGET = 1 and taking equal number of data points 
#which will balance the dataset 
def balanced_df_set(dff1):
    df_target_1 = dff1[df1['TARGET'] == 1]
    df_target_0 = dff1[df1['TARGET'] == 0]
    size_target_1= len(df_target_1)
    df_target_0_sample= df_target_0.sample(n=size_target_1)
    df_sample = df_target_0_sample.append(df_target_1)
    return df_sample

```

```python
#function takes in a sample dataframe with a balanced data and find the correlation between them and 'TARGET'. The for loop 
# with a value of 21 signifies that it has itrated 20 times to find the top 20 columns 
def corelator_generator(df_balance):
    corr_df=df_balance.corr()
    corr_df = corr_df.sort_values(by ='TARGET')
    Top_20_corr_df_positive = (corr_df.nlargest(21,['TARGET']))
    Top_20_corr_df_negative = (corr_df.nsmallest(21,['TARGET']))*(-1)
    Top_40_corr_df= Top_20_corr_df_positive
    Top_40_corr_df = Top_40_corr_df.append(Top_20_corr_df_negative)
    final_top_20_corr_df = (Top_40_corr_df.nlargest(21,['TARGET']))
    top_20_columns = final_top_20_corr_df.index.values
    return top_20_columns
```

```python
# this function is used to call all the above defined function and return final top columns which has high correlation with
#target
import operator
def top_correlated_cols(dff1):
    top_cols= {}
    df_all= convert_cat_noncat(dff1)
    for x in range(0,21):
        df_balance = balanced_df_set(df_all)
        top_20_columns= corelator_generator(df_balance)
        for x in top_20_columns:
            if x !='TARGET':
                if x in top_cols:
                    top_cols[x]=top_cols[x]+1
                else:
                    top_cols[x]=1
    sorted_top_cols = sorted(top_cols.items(),  key=operator.itemgetter(1), reverse= True)
    return sorted_top_cols         
    
```

```python
#creating dataframe to the graph
sorted_top_cols=top_correlated_cols(dff1)
dff2= pd.DataFrame()
#sorted_top_cols=top_correlated_cols(dff1)
for x in range (0,20):
    key =sorted_top_cols[x]
    dff2=pd.concat([dff2,dff1[key[0]]],axis=1)
final_corr = dff2.corr()

```

```python
#to find top 10 independent categorical variables with correlation with Target
def top_10_independent_corr(final_corr):
    for x in final_corr:
        for y in final_corr:
            if final_corr[x][y]<0:
                final_corr[x][y] = final_corr[x][y]*(-1)
        final_corr[x][x]=-1
    for x in range(0,10):
        max=-1
        final_y=''
        final_x=''
        for x in final_corr:
            for y in final_corr:
                if final_corr[x][y]>max:
                    max= final_corr[x][y]
                    final_x=x
                    final_y=y
        final_corr[final_x][final_y]=-1
        final_corr[final_y][final_x]=-1
    top_20_corr= list(final_corr.columns.values)    
    #top_20_corr.remove('EXT_SOURCE_3')
    for x in final_corr:
        for y in final_corr:
            if final_corr[x][y]==-1 and x!=y:
                if x in top_20_corr:
                    top_20_corr.remove(x)
                    if len(top_20_corr)<=10:
                        return top_20_corr

                    
                    
          
```
